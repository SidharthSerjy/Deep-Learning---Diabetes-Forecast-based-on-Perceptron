## Baseline Perceptron Model

```python


# Perceptron implementation
class Perceptron:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.activation_func = self._step_function
        self.weights = None
        self.bias = None

    def _step_function(self, x):
        return np.where(x >= 0, 1, -1)

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)


                update = self.lr * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted

# Prepare the data for the perceptron model
X = df_cleaned.drop(columns=['Diabetes_Outcome']).astype(float).values
y = df_cleaned['Diabetes_Outcome'].replace({'+1': 1, '-1': -1}).values


perceptron = Perceptron(learning_rate=0.01, n_iters=1000)


perceptron.fit(X, y)


predictions = perceptron.predict(X)


accuracy = np.mean(predictions == y)
accuracy
```

Experimenting with early stoping on the sigmoid function

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define squared loss
def squared_loss(y_true, y_pred):
    return (y_true - y_pred) ** 2

class Perceptron:
    def __init__(self, learning_rate=0.01, n_iters=1000, early_stopping=False, tolerance=1e-4):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.early_stopping = early_stopping
        self.tolerance = tolerance
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        best_weights = None
        best_bias = None
        best_accuracy = 0
        no_improvement_count = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = sigmoid(linear_output)

                # Perceptron update rule with squared loss
                loss = squared_loss(y[idx], y_predicted)
                update = self.lr * loss * (y[idx] - y_predicted)

                # Update weights and bias
                self.weights += update * x_i
                self.bias += update

            # Early stopping logic
            if self.early_stopping:
                y_pred = self.predict(X)
                accuracy = np.mean(y_pred == y)

                if accuracy > best_accuracy + self.tolerance:
                    best_accuracy = accuracy
                    best_weights = self.weights.copy()
                    best_bias = self.bias
                    no_improvement_count = 0
                else:
                    no_improvement_count += 1

                if no_improvement_count >= 10:  # Stop if no improvement over 10 iterations
                    print("Early stopping triggered")
                    break


        if self.early_stopping:
            self.weights = best_weights
            self.bias = best_bias

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = sigmoid(linear_output)  # Using sigmoid activation
        return np.where(y_predicted >= 0.5, 1, -1)  # Convert back to binary classes



# Initialize Perceptron model with Sigmoid activation and early stopping
perceptron_sigmoid_early_stopping = Perceptron(learning_rate=0.01, n_iters=1000, early_stopping=True, tolerance=1e-4)


perceptron_sigmoid_early_stopping.fit(X, y)


predictions_sigmoid_early_stopping = perceptron_sigmoid_early_stopping.predict(X)


accuracy_sigmoid_early_stopping = np.mean(predictions_sigmoid_early_stopping == y)
print(f"Accuracy with sigmoid activation and early stopping: {accuracy_sigmoid_early_stopping:.4f}")
```

Experimenting with the hinge loss on step wise function

```python
import numpy as np

# Define hinge loss function
def hinge_loss(y_true, y_pred):
    return np.maximum(0, 1 - y_true * y_pred)

class Perceptron:
    def __init__(self, learning_rate=0.01, n_iters=1000, regularization='l2', reg_lambda=0.01):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.regularization = regularization  # 'l1' or 'l2'
        self.reg_lambda = reg_lambda  # Regularization strength
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = np.sign(linear_output)

                # Compute hinge loss
                loss = hinge_loss(y[idx], y_predicted)

                # Perceptron update rule with hinge loss and regularization
                update = self.lr * loss * y[idx]

                # Update weights with regularization (L1 or L2)
                if self.regularization == 'l2':
                    self.weights += update * x_i - self.reg_lambda * self.weights  # L2 regularization
                elif self.regularization == 'l1':
                    self.weights += update * x_i - self.reg_lambda * np.sign(self.weights)  # L1 regularization

                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return np.where(linear_output >= 0, 1, -1)  # Step function output for binary classification



# Initialize Perceptron model with Hinge Loss and L2 regularization
perceptron_hinge_l2 = Perceptron(learning_rate=0.01, n_iters=1000, regularization='l2', reg_lambda=0.01)

# Train the model
perceptron_hinge_l2.fit(X, y)


predictions_hinge_l2 = perceptron_hinge_l2.predict(X)


accuracy_hinge_l2 = np.mean(predictions_hinge_l2 == y)
print(f"Accuracy with hinge loss and L2 regularization: {accuracy_hinge_l2:.4f}")


perceptron_hinge_l1 = Perceptron(learning_rate=0.01, n_iters=1000, regularization='l1', reg_lambda=0.01)

# Train the model with L1 regularization
perceptron_hinge_l1.fit(X, y)


predictions_hinge_l1 = perceptron_hinge_l1.predict(X)


accuracy_hinge_l1 = np.mean(predictions_hinge_l1 == y)
print(f"Accuracy with hinge loss and L1 regularization: {accuracy_hinge_l1:.4f}")
```
